```python
# Importations nécessaires
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
from transformers import AutoTokenizer, GPT2LMHeadModel, AdamW
from datasets import load_dataset, DatasetDict
from tqdm import tqdm
import random
import numpy as np
import os
```

```python
# Paramètres globaux
SEED = 42
MAX_LEN = 512
BATCH_SIZE = 8
EPOCHS = 3
LEARNING_RATE = 5e-5
MODEL_NAME = "aubmindlab/aragpt2-base"

# Reproductibilité
torch.manual_seed(SEED)
random.seed(SEED)
np.random.seed(SEED)
```

```python
# Charger les datasets avec Hugging Face
raw_datasets = DatasetDict({
    "train": load_dataset("json", data_files="train-open.json", split="train"),
    "validation": load_dataset("json", data_files="val-open.json", split="train"),
    "test": load_dataset("json", data_files="test-open.json", split="train"),
})

# Fonction de prétraitement
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token  # Ajout du token PAD

def preprocess_function(examples):
    input_texts = [f"Question: {q}\nAnswer:" for q in examples['question']]
    target_texts = examples['answer']

    model_inputs = tokenizer(
        input_texts, max_length=MAX_LEN, padding="max_length", truncation=True
    )
    labels = tokenizer(
        target_texts, max_length=MAX_LEN, padding="max_length", truncation=True
    )['input_ids']
    model_inputs["labels"] = labels
    return model_inputs

# Appliquer la tokenization
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
tokenized_datasets.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
# Création des DataLoaders
train_dataloader = DataLoader(tokenized_datasets["train"], batch_size=BATCH_SIZE, shuffle=True)
val_dataloader = DataLoader(tokenized_datasets["validation"], batch_size=BATCH_SIZE)
test_dataloader = DataLoader(tokenized_datasets["test"], batch_size=BATCH_SIZE)
```

```python
class DifferentialAttention(nn.Module):
    def __init__(self, d_model, num_heads, lambda_init=0.8):
        super().__init__()
        self.num_heads = num_heads
        self.d = d_model // (2 * num_heads)
        self.lambda_init = lambda_init
        self.WQ = nn.Linear(d_model, 2 * d_model)
        self.WK = nn.Linear(d_model, 2 * d_model)
        self.WV = nn.Linear(d_model, d_model)
        self.lambda_q1 = nn.Parameter(torch.randn(self.d, 1))

    def forward(self, X):
        Q = self.WQ(X)
        K = self.WK(X)
        V = self.WV(X)
        Q1, Q2 = Q.chunk(2, dim=-1)
        K1, K2 = K.chunk(2, dim=-1)

        lambda_val = torch.exp(self.lambda_q1.T @ self.lambda_q1).squeeze()
        attn1 = F.softmax(torch.matmul(Q1, K1.transpose(-1, -2)) / (self.d ** 0.5), dim=-1)
        attn2 = F.softmax(torch.matmul(Q2, K2.transpose(-1, -2)) / (self.d ** 0.5), dim=-1)
        output = (attn1 - lambda_val * attn2) @ V
        return output

class GPT2WithDifferentialAttention(nn.Module):
    def __init__(self, model_name):
        super().__init__()
        self.gpt2 = GPT2LMHeadModel.from_pretrained(model_name)
        self.diff_attention = DifferentialAttention(d_model=768, num_heads=12)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.gpt2.transformer(input_ids=input_ids, attention_mask=attention_mask)
        diff_output = self.diff_attention(outputs.last_hidden_state)
        logits = self.gpt2.lm_head(diff_output)
        loss = None
        if labels is not None:
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), labels.view(-1), ignore_index=-100)
        return loss, logits
```

```python
device = "cpu"  # Exécution sur CPU
model = GPT2WithDifferentialAttention(MODEL_NAME).to(device)
optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

# Boucle d'entraînement
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for batch in tqdm(train_dataloader, desc=f"Training Epoch {epoch + 1}"):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        loss, _ = model(input_ids, attention_mask, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    avg_loss = total_loss / len(train_dataloader)
    print(f"Epoch {epoch + 1} | Average Loss: {avg_loss:.4f}")
```

```python
model.eval()
total_eval_loss = 0

for batch in tqdm(val_dataloader, desc="Evaluating"):
    input_ids = batch['input_ids'].to(device)
    attention_mask = batch['attention_mask'].to(device)
    labels = batch['labels'].to(device)

    with torch.no_grad():
        loss, _ = model(input_ids, attention_mask, labels)
        total_eval_loss += loss.item()

avg_val_loss = total_eval_loss / len(val_dataloader)
print(f"Validation Loss: {avg_val_loss:.4f}")
```

```python
# Sauvegarder les poids du modèle fine-tuné
model_save_path = "fine_tuned_aragpt2_diff_attention_cpu.pt"
torch.save(model.state_dict(), model_save_path)
print(f"Modèle sauvegardé avec succès : {model_save_path}")
```

```python
model.eval()
predictions = []

for batch in tqdm(test_dataloader, desc="Testing"):
    input_ids = batch['input_ids'].to(device)
    attention_mask = batch['attention_mask'].to(device)

    with torch.no_grad():
        _, logits = model(input_ids, attention_mask)
        preds = torch.argmax(logits, dim=-1)
        predictions.extend(preds.cpu().numpy())

print("Prédictions générées avec succès sur le jeu de test.")
```
