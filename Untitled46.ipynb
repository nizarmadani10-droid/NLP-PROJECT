```python
import pandas as pd
from datasets import Dataset
```

```python
train_data = pd.read_json("train-open.json", lines=False)
val_data = pd.read_json("val-open.json", lines=False)
test_data = pd.read_json("test-open.json", lines=False)

# If column names differ, rename them
train_data.rename(columns={"input_question": "question", "response": "answer"}, inplace=True)
val_data.rename(columns={"input_question": "question", "response": "answer"}, inplace=True)
test_data.rename(columns={"input_question": "question", "response": "answer"}, inplace=True)

# Check and clean missing data
train_data.dropna(subset=["question", "answer"], inplace=True)
val_data.dropna(subset=["question", "answer"], inplace=True)
test_data.dropna(subset=["question", "answer"], inplace=True)
```

```python
def augment_data(row):
    return [
        {"question": row["question"] + "?", "answer": row["answer"]},  # Add question mark
        {"question": row["question"].upper(), "answer": row["answer"]},  # Uppercase variation
    ]
```

```python
augmented_train_data = train_data.apply(augment_data, axis=1).explode().reset_index(drop=True)
augmented_train_data = pd.DataFrame(augmented_train_data.tolist())
```

```python
train_dataset = Dataset.from_pandas(augmented_train_data)
val_dataset = Dataset.from_pandas(val_data)
test_dataset = Dataset.from_pandas(test_data)

# Subsample the datasets for faster development (use full dataset for final training)
train_dataset = train_dataset.select(range(min(len(train_dataset), 10000)))  # Up to 10k examples
val_dataset = val_dataset.select(range(min(len(val_dataset), 2000)))         # Up to 2k examples
test_dataset = test_dataset.select(range(min(len(test_dataset), 2000)))     # Up to 2k examples
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/aragpt2-base")
tokenizer.pad_token = tokenizer.eos_token  # Set pad_token to eos_token
```

```python
def preprocess_function(examples):
    inputs = examples["question"]
    targets = examples["answer"]

    model_inputs = tokenizer(
        inputs, max_length=128, truncation=True, padding="max_length"
    )
    labels = tokenizer(
        targets, max_length=128, truncation=True, padding="max_length"
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)
tokenized_val_dataset = val_dataset.map(preprocess_function, batched=True)
tokenized_test_dataset = test_dataset.map(preprocess_function, batched=True)
```

```python
from datasets import DatasetDict

tokenized_datasets = DatasetDict({
    "train": tokenized_train_dataset,
    "validation": tokenized_val_dataset,
    "test": tokenized_test_dataset,
})
```

```python
from transformers import AutoModelForCausalLM

# Load the model
model = AutoModelForCausalLM.from_pretrained("aubmindlab/aragpt2-base")
optimal_vocab_size = (len(tokenizer) + 7) // 8 * 8
model.resize_token_embeddings(optimal_vocab_size)
```

```python
import torch
for module in model.modules():
    if isinstance(module, torch.nn.Dropout):
        module.p = 0.3  # Increase dropout to 30%

# Define Optimizers
from torch.optim import AdamW, SGD, RMSprop

adamw_optimizer = AdamW(model.parameters(), lr=5e-5, weight_decay=0.01)
sgd_optimizer = SGD(model.parameters(), lr=5e-5, momentum=0.9)
rmsprop_optimizer = RMSprop(model.parameters(), lr=5e-5, alpha=0.99)
```

```python
from transformers import TrainingArguments, Trainer
from transformers import DataCollatorWithPadding
from sklearn.metrics import accuracy_score, f1_score
from torchmetrics.text.bleu import BLEUScore
import matplotlib.pyplot as plt
```

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, padding=True)
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    acc = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average="weighted")
    bleu = BLEUScore()
    return {"accuracy": acc, "f1_score": f1, "bleu_score": bleu(predictions, labels)}
```

```python
training_args = TrainingArguments(
    output_dir="./base_model",
    evaluation_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    save_steps=1000,
    save_total_limit=2,
    logging_dir="./logs_base",
    logging_steps=50,
    fp16=True,
    gradient_accumulation_steps=2,
    dataloader_num_workers=2,
)
```

```python
trainer_base = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
trainer_base.train()
base_results = trainer_base.evaluate(tokenized_datasets["test"])
print("Base Model Results:", base_results)
```

```python
import torch
from torch import nn
import math

class MultiHeadDifferentialAttention(nn.Module):
    def __init__(self, d_model, num_heads, lambda_init=0.8):
        super(MultiHeadDifferentialAttention, self).__init__()
        self.num_heads = num_heads
        self.d_head = d_model // num_heads
        self.scale = 1 / math.sqrt(self.d_head)

        # Linear projections for Q, K, and V
        self.w_q = nn.Linear(d_model, d_model * 2)  # Q1 and Q2 combined
        self.w_k = nn.Linear(d_model, d_model * 2)  # K1 and K2 combined
        self.w_v = nn.Linear(d_model, d_model)     # V projection

        # Learnable lambda components
        self.lambda_q1 = nn.Parameter(torch.zeros(self.d_head))
        self.lambda_k1 = nn.Parameter(torch.zeros(self.d_head))
        self.lambda_q2 = nn.Parameter(torch.zeros(self.d_head))
        self.lambda_k2 = nn.Parameter(torch.zeros(self.d_head))
        self.lambda_init = lambda_init

    def forward(self, hidden_states, layer_past=None, attention_mask=None, **kwargs):
        # Project inputs to Q, K, and V
        q1, q2 = self.w_q(hidden_states).chunk(2, dim=-1)  # Split into Q1 and Q2
        k1, k2 = self.w_k(hidden_states).chunk(2, dim=-1)  # Split into K1 and K2
        v = self.w_v(hidden_states)                       # V projection

        # Compute scaled attention scores
        attn1 = torch.softmax((q1 @ k1.transpose(-2, -1)) * self.scale, dim=-1)
        attn2 = torch.softmax((q2 @ k2.transpose(-2, -1)) * self.scale, dim=-1)

        # Learnable lambda calculation
        lambda_value = (
            torch.exp(self.lambda_q1 @ self.lambda_k1)  # Positive scaling
            - torch.exp(self.lambda_q2 @ self.lambda_k2)  # Negative scaling
            + self.lambda_init                           # Initialization value
        )

        # Compute differential attention
        diff_attn = attn1 - lambda_value * attn2
        updated_hidden_states = diff_attn @ v

        # Return updated hidden states, None for past key values, and None for attention weights
        return (updated_hidden_states, None, None)
```

```python
from transformers.models.gpt2.modeling_gpt2 import GPT2Attention

class ModifiedGPT2Attention(nn.Module):
    def __init__(self, config, use_differential_attention=False):
        super(ModifiedGPT2Attention, self).__init__()
        self.use_differential_attention = use_differential_attention

        if use_differential_attention:
            # Use Differential Attention
            self.attention = MultiHeadDifferentialAttention(
                d_model=config.hidden_size,
                num_heads=config.num_attention_heads
            )
        else:
            # Use default GPT-2 attention
            self.attention = GPT2Attention(config)

    def forward(self, hidden_states, layer_past=None, attention_mask=None, **kwargs):
        if self.use_differential_attention:
            # Pass inputs to differential attention
            return self.attention(hidden_states, layer_past=layer_past, attention_mask=attention_mask, **kwargs)
        else:
            # Default GPT-2 attention
            return self.attention(hidden_states, layer_past=layer_past, attention_mask=attention_mask, **kwargs)
```

```python
from transformers import AutoModelForCausalLM

def modify_attention_layers(model, percentage_replaced=0.25):
    num_layers = len(model.transformer.h)
    num_replacements = int(num_layers * percentage_replaced)

    for i in range(num_replacements):
        model.transformer.h[i].attn = ModifiedGPT2Attention(
            model.config, use_differential_attention=True
        )
    return model

# Load the model and apply modifications
upgraded_model = AutoModelForCausalLM.from_pretrained("aubmindlab/aragpt2-base")
upgraded_model.resize_token_embeddings(optimal_vocab_size)
upgraded_model = modify_attention_layers(upgraded_model, percentage_replaced=0.25)
```

```python
# Training arguments for the upgraded model
training_args_upgraded = TrainingArguments(
    output_dir="./upgraded_model",
    evaluation_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    save_steps=1000,
    save_total_limit=2,
    logging_dir="./logs_upgraded",
    logging_steps=50,
    fp16=True,
    gradient_accumulation_steps=2,
    dataloader_num_workers=2,
)
```

```python
trainer_upgraded = Trainer(
    model=upgraded_model,
    args=training_args_upgraded,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
trainer_upgraded.train()
upgraded_results = trainer_upgraded.evaluate(tokenized_datasets["test"])
print("Upgraded Model Results:", upgraded_results)
```

```python
# Compare base and upgraded models' metrics
metrics = ["accuracy", "f1_score", "bleu_score"]
base_scores = [base_results[m] for m in metrics]
upgraded_scores = [upgraded_results[m] for m in metrics]

# Plot metrics
x = range(len(metrics))
plt.figure(figsize=(10, 6))
plt.bar(x, base_scores, width=0.4, label="Base Model", align="center")
plt.bar([p + 0.4 for p in x], upgraded_scores, width=0.4, label="Upgraded Model", align="center")
plt.xticks([p + 0.2 for p in x], metrics)
plt.ylabel("Score")
plt.title("Metrics Comparison: Base vs Upgraded Model")
plt.legend()
plt.grid(axis="y")
plt.show()
```
